```python
import json

# Đối tượng Python (đối tượng dict)
data = {
    "Ho_va_ten": "Nguyễn Văn A",
    "Nam_sinh": 2005,
    "Gioi_tinh": "Nam",
    "so_dt": "0123456789",
    "email": "nguyenvana@uneti.edu.vn"
}

# Sử dụng 'with open' để mở tệp 'thongtin.json' với chế độ ghi ('w') và mã hóa utf-8
# Hàm json.dump() sẽ ghi đối tượng Python trực tiếp vào tệp đã mở.
with open("thongtin.json", "w", encoding="utf-8") as f:          # Mở tệp thongtin.json để ghi dữ liệu, nếu tệp không tồn tại sẽ tự động tạo mới
    json.dump(data, f, ensure_ascii=False, indent=4)             # Hàm dump() ghi đối tượng Python vào file thongtin.json
    print("Dữ liệu đã được ghi thành công vào tệp 'thongtin.json'.")

# (Tùy chọn) In nội dung file để kiểm tra
with open("thongtin.json", "r", encoding="utf-8") as f:
    print("\nNội dung tệp 'thongtin.json':")
    print(f.read())
```

```python
import json

# Hàm kiểm tra chuỗi JSON có chứa đối tượng phức (complex object) hay không
def is_complex(object):
    # Kiểm tra xem trong đối tượng có tồn tại khóa "__complex__" hay không
    if "__complex__" in object:
        # Nếu có, tạo một số phức từ thông tin trong đối tượng
        return complex(object['real'], object['img'])
    # Nếu không có khóa "__complex__", tức không phải là đối tượng phức, trả về đối tượng gốc
    return object

def main():
    # Sử dụng hàm json.loads() với object_hook để kiểm tra và chuyển đổi đối tượng phức trong chuỗi JSON
    complex_object = json.loads('{"__complex__": true, "real": 4, "img": 5}', object_hook=is_complex)  # Đối số object_hook chỉ định hàm is_complex để xử lý các đối tượng khi phân tích
    simple_object = json.loads('{"real": 6, "img": 7}', object_hook=is_complex)

    # In kết quả phân tích chuỗi JSON cho cả đối tượng phức và đối tượng thường
    print("Đối tượng số phức:", complex_object)
    print("Đối tượng bình thường:", simple_object)

if __name__ == "__main__":
    main()
```

```python
import json
# Import lớp JSONDecoder từ module json.decoder
from json.decoder import JSONDecoder

# Chuỗi JSON chứa dữ liệu màu sắc.
colour_string = '{ "colour": ["red", "yellow"]}'

# Sử dụng JSONDecoder để giải mã chuỗi JSON thành đối tượng Python.
# JSONDecoder là một lớp con của json.JSONDecoder, cho phép tùy chỉnh quá trình giải mã.
# Phương thức decode() sẽ giải mã chuỗi JSON và trả về đối tượng Python tương ứng.
decoded_object = JSONDecoder().decode(colour_string)

# In đối tượng Python đã giải mã và kiểu của đối tượng đó ra màn hình.
print(decoded_object)
print(type(decoded_object))
```

```python
import requests

# Thực hiện yêu cầu HTTP đến URL 'https://jsonplaceholder.typicode.com/posts'
response = requests.get('https://jsonplaceholder.typicode.com/posts')

# Kiểm tra xem yêu cầu thành công hay không
if response.status_code == 200:
    # Nếu yêu cầu thành công, lấy dữ liệu JSON từ phản hồi
    json_data = response.json()

    # In ra màn hình thông tin các bài đăng từ dữ liệu JSON
    for post in json_data:
        print("ID bài đăng:", post['id'])
        print("Tiêu đề:", post['title'])
        print("Nội dung:", post['body'])
        print("=======================================")
else:
    print("Yêu cầu không thành công. Mã trạng thái:", response.status_code)
```
