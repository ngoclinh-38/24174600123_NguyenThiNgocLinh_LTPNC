```python
#ví dụ minh hoạ đối số obj và ensure_ascii 
import json

data = {"name": "Ng Văn A", "age": 23, "skills": ["AI", "Python"]}  #data là đối tượng Python kiểu dữ liệu từ điển (dict)
json_st1 = json.dumps(data, ensure_ascii=False)  #chuyển đổi đối tượng Python thành chuỗi JSON, đối số ensure_ascii=False để giữ nguyên ký tự
print(json_st1)
```

```python
#ví dụ minh hoạ đối số indent
json_st2 = json.dumps(data, indent=4, ensure_ascii=False)  # chuyển đổi đối tượng Python thành chuỗi JSON với định dạng thụt lề 4 khoảng
print(json_st2)
```

```python
#Đối số default
import json

def complex_handler(obj):                      # Hàm xử lý số phức kiểm tra nếu obj là số phức thì trả về danh sách gồm phần thực và phần ảo
    if isinstance(obj, complex):
        return [obj.real, obj.imag]            # Trả về danh sách gồm phần thực và phần ảo của số phức
    raise TypeError(repr(obj) + " không thể mã hóa JSON")   # Nếu không phải số phức, ném ra lỗi TypeError

# Ứng dụng hàm complex_handler để mã hóa số phức
data = {"z": 2 + 3j}
complex_obj = json.dumps(data, default=complex_handler)     # Sử dụng đối số default để chỉ định hàm xử lý tùy chỉnh complex_handler()
print(complex_obj)
```
