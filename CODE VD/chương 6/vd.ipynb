```python
import threading
def print_cube(num):
    # function to print cube of given num
    print("Cube: {}".format(num * num * num))
def print_square(num):
    # function to print square of given num
    print("Square: {}".format(num * num))
if __name__ == "__main__":
    # creating thread
    t1 = threading.Thread(target=print_square, args=(10,))
    t2 = threading.Thread(target=print_cube, args=(10,))
    # start the threads
    t1.start()
    t2.start()
    # wait until both threads finish
    t1.join()
    t2.join()

    # continue with the main loop
    for i in range(5):
        print("Main loop: {}".format(i))
    print("Done!")
```

```python
import threading

class PrintThread(threading.Thread):
    def __init__(self, num):
        super().__init__()
        self.num = num

    def run(self):
        self.print_square()
        self.print_cube()

    def print_cube(self):
        # function to print cube of given num
        print("Cube: {}".format(self.num * self.num * self.num))

    def print_square(self):
        # function to print square of given num
        print("Square: {}".format(self.num * self.num))

if __name__ == "__main__":
    # creating thread
    t = PrintThread(10)

    # start the thread
    t.start()

    # wait until the thread finishes
    t.join()

    # continue with the main loop
    for i in range(5):
        print("Main loop: {}".format(i))
    print("Done!")
```

```python
# Chương trình minh họa khái niệm luồng 'Threading'
import threading
import os

def task1():
    print("Gắn tác vụ 1 đến luồng: {}".format(threading.current_thread().name))
    print("ID của tiến trình chạy task 1: {}".format(os.getpid()))

def task2():
    print("Gắn tác vụ 1 đến luồng: {}".format(threading.current_thread().name))
    print("ID của tiến trình chạy task 2: {}".format(os.getpid()))

if __name__ == "__main__":
    # In ra ID của luồng hiện thời
    print("ID của luồng đang thực thi trong chương trình chính: {}".format(os.getpid()))

    # in ra tên luồng chính
    print("Tên của luồng chính: {}".format(threading.current_thread().name))

    # Tạo các luồng
    t1 = threading.Thread(target=task1, name='t1')
    t2 = threading.Thread(target=task2, name='t2')

    # Bắt đầu các luồng:
    t1.start()
    t2.start()
    
    # Chờ cho đến khi các luồng kết thúc:
    t1.join()
    t2.join()
```

```python
#A) Đọc file tuần tự (single-thread)
import time                           # Để tính thời gian thực thi
import pandas as pd

files = ["sales_jan.csv", "sales_feb.csv", "sales_mar.csv"]

def read_file_sequential():
    start = time.time()

    for f in files:
        time.sleep(0.2)
        df = pd.read_csv(f)
        print(f"Đọc xong {f}: {len(df)} dòng")

    end = time.time()
    print(f"Thời gian đọc tuần tự: {end - start:.4f} giây")
```

```python
read_file_sequential()
```

```python
#B) Đọc file bằng threading.Thread
import threading
import time
import pandas as pd

results = []

def load_csv_thread(filename):
    time.sleep(0.2)
    df = pd.read_csv(filename)
    results.append((filename, len(df)))
    print(f"Đọc xong {filename}: {len(df)} dòng")

def read_file_multithread():
    start = time.time()

    threads = [] #Danh sách lưu trữ các luồng
    for f in files:                                             #Lặp qua danh sách files cần đọc sales_feb, sales_mar, sales_jan
        t = threading.Thread(target=load_csv_thread, args=(f,)) #Tạo luồng mới bằng cách truyền hàm load_csv_thread và đối số f
        threads.append(t) #Thêm luồng vào danh sách lưu trữ threads
        t.start() #Bắt đầu luồng ở mỗi lần lặp

    for t in threads:
        t.join() #Chờ tất cả các luồng kết thúc

    end = time.time()
    print(f"Thời gian đọc đa luồng: {end - start:.4f} giây") #In ra thời gian thực thi tổng cộng khi sử dụng đa luồng
```

```python
read_file_multithread()
```

```python
#C) Hàm so sánh trực tiếp
if __name__ == "__main__":
    print("——— Đọc tuần tự ——")
    read_file_sequential()

    print("\n——— Đọc đa luồng ——")
    read_file_multithread()
```

```python
from concurrent.futures import ThreadPoolExecutor
import pandas as pd
import time

def load_csv(filename):
    print(f"Đang đọc {filename}")
    time.sleep(0.5)
    df = pd.read_csv(filename)
    return filename, len(df)

files = ["sales_jan.csv", "sales_feb.csv", "sales_mar.csv"]

with ThreadPoolExecutor(max_workers=3) as executor: # Tạo một bộ gồm tối đa ba luồng, tương ứng với số lượng file trong danh sách.
    results = executor.map(load_csv, files)       # Lần lượt gửi từng tên file vào hàm load_csv.
                                                   # Tại thời điểm này, ba luồng được kích hoạt gần như đồng thời, mỗi luồng đọc một file CSV.

for fname, rows in results:                        # Kết quả trả về được gom lại theo đúng thứ tự danh sách input
    print(f"{fname}: {rows} dòng")
```

```python
import threading

# global variable x
x = 0

# Viết hàm increment(), tăng giá trị biến toàn cục x
def increment():
    global x
    x += 1

# Hàm thread_task() thực hiện tác vụ luồng, thực hiện gọi hàm increment() 100000 lần...
def thread_task():
    for _ in range(100000):
        increment()

# Hàm main_task(): thực hiện nhiệm vụ chính của chương trình.
# Nó đặt giá trị của biến toàn cục x=0, tạo hai luồng (t1 và t2)
# để thực thi công việc trong thread_task().
def main_task():
    global x
    x = 0

    # creating threads
    t1 = threading.Thread(target=thread_task)
    t2 = threading.Thread(target=thread_task)

    # start threads
    t1.start()
    t2.start()

    # wait until threads finish their job
    t1.join()
    t2.join()

# Chương trình chạy vòng lặp 10 lần,
# mỗi lần gọi hàm main_task() để thực hiện công việc.
# Sau đó in giá trị x mỗi lần lặp ra màn hình.
for i in range(10):
    main_task()
    print("Lần lặp {0}: x = {1}".format(i, x))
```

```python
import threading
import time

# global variable x
x = 0

# Viết hàm increment(), tăng giá trị biến toàn cục x
def increment():
    global x
    tmp = x
    time.sleep(0.0000001)  # Ép thread switching
    x = tmp + 1

# Hàm thread_task() thực hiện tác vụ luồng, thực hiện gọi hàm increment() 100000 lần...
def thread_task():
    for _ in range(100000):
        increment()

# Hàm main_task(): thực hiện nhiệm vụ chính của chương trình.
# Nó đặt giá trị của biến toàn cục x=0, tạo hai luồng (t1 và t2)
# để thực thi công việc trong thread_task().
def main_task():
    global x
    x = 0

    # creating threads
    t1 = threading.Thread(target=thread_task)
    t2 = threading.Thread(target=thread_task)

    # start threads
    t1.start()
    t2.start()

    # wait until threads finish their job
    t1.join()
    t2.join()

# Chương trình chạy vòng lặp 10 lần,
# mỗi lần gọi hàm main_task() để thực hiện công việc.
# Sau đó in giá trị x mỗi lần lặp ra màn hình.
for i in range(10):
    main_task()
    print("Lần lặp {0}: x = {1}".format(i, x))
```

```python
import threading
import time

balance = 100        # Biến toàn cục, được CHIA SẺ giữa các thread

def add_money():
    global balance
    tmp = balance          # Bước 1: đọc số dư hiện tại vào biến tạm
    time.sleep(0.001)      # Bước 2: Ngủ 1ms -> có tính tạo cơ hội cho thread khác nhảy vào (ép thread switching)
    balance = tmp + 10     # Bước 3: Dùng giá trị ĐÃ ĐỌC từ rất lâu trước đó để ghi lại

# Tiếp theo: tạo 2 thread, cả 2 đều chạy cùng một hàm add_money()
threads = [threading.Thread(target=add_money) for _ in range(2)] #Threads là nơi chứa các luồng được tạo ra

for t in threads:            # Cả 2 thread bắt đầu chạy gần như cùng lúc
    t.start()

for t in threads:            # Đợi cả 2 thread chạy xong
    t.join()

print("Số dư cuối:", balance)
```

```python
import threading
import time

#Tạo 1 lock
lock = threading.Lock()
balance = 100        # Biến toàn cục, được CHIA SẺ giữa các thread

def add_money_safe():
    global balance
    #Yêu cầu khoá. Chỉ 1 luồng đc đi qua.
    lock.acquire()
    try:    
        tmp = balance          # Bước 1: đọc số dư hiện tại vào biến tạm
        time.sleep(0.001)      # Bước 2: Ngủ 1ms -> có tính tạo cơ hội cho thread khác nhảy vào (ép thread switching)
        balance = tmp + 10     # Bước 3: Dùng giá trị ĐÃ ĐỌC từ rất lâu trước đó để ghi lại
    finally:
        #Giải phóng khoá
        lock.release()

# Tiếp theo: tạo 2 thread, cả 2 đều chạy cùng một hàm add_money()
threads = [threading.Thread(target=add_money) for _ in range(2)] #Threads là nơi chứa các luồng được tạo ra

for t in threads:            # Cả 2 thread bắt đầu chạy gần như cùng lúc
    t.start()

for t in threads:            # Đợi cả 2 thread chạy xong
    t.join()

print("Số dư cuối:", balance)
```

```python
import threading
import time

# Khởi tạo Semaphore với giá trị 3
sem = threading.Semaphore(3)

def access(id):
    # Luồng cố gắng lấy "quyền" truy cập
    sem.acquire()
    
    # Khu vực truy cập giới hạn
    print(f"Thread {id} bắt đầu truy cập")
    time.sleep(1) # Mô phỏng thời gian xử lý
    print(f"Thread {id} kết thúc")
    # Luồng trả lại "quyền" truy cập
    sem.release()

# Khởi tạo và bắt đầu 6 luồng
for i in range(6):
    threading.Thread(target=access, args=(i,)).start()
```
