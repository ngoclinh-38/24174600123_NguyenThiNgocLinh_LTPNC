```python
import numpy as np
import pandas as pd

#1. None khiến dtype chuyển thành object
arr_none = np.array([1, 2, None, 3])
print("1. Mảng có None (dtype=object):")
print(arr_none)
```

```python
type(arr_none)
```

```python
#2. Minh hoạ lỗi phép toán
print("\n2. Lỗi khi tính mean() trên mảng object chứa None:")
try: 
    #Phép toán số học k thực hiện đc trên object chứa None
    arr_none.mean()
except TypeError as e:
    print("Lỗi:", e)
```

```python
print("\n3. Minh hoạ lỗi khi tính sum() trên object thuần:")
ar_pure_obj = np.array([1, 2, 3, 4], dtype=object)
try:
    #Ngay cả khi không có None, phép toán số học vẫn k thực hiện đc trên object thuần
    ar_pure_obj.sum()
except TypeError as e:
    print(f" -> Lỗi: {e} (do có chuỗi '4')")
```

```python
#3. So sánh tốc độ tính toán giữa mảng dtype object và int
for dtype in ['object', 'int']:
    print("dtype:", dtype)
    %timeit np.arange(1e6, dtype=dtype).sum()
    print()
```

```python
import pandas as pd

# #1. Định nghĩa Tuple đại diện cho tỉnh/TP - Năm tương ứng
index_tuples = [
    ('Hà Nội', 2015), ('Hà Nội', 2020),
    ('Nam Định', 2015), ('Nam Định', 2020),
    ('Thanh Hóa', 2015), ('Thanh Hóa', 2020),
    ('Đà Nẵng', 2015), ('Đà Nẵng', 2020),
    ('Hồ Chí Minh', 2015), ('Hồ Chí Minh', 2020)
]

# #2. Dữ liệu (Sử dụng nguồn dữ liệu chính xác đã thống nhất)
data = [
    7216000, 8246540,
    1850600, 1780330,
    3514200, 3664940,
    1028800, 1169480,
    8146300, 9227600
]

# Sau bước này, bạn sẽ tạo Series hoặc DataFrame bằng cách sử dụng
# index=index_tuples và data=data, ví dụ:
# vn_pop_2015_2020 = pd.Series(data=data, index=index_tuples)

# 3. Tạo Series (CHỈ SỬ DỤNG TUPLE INDEX)
vn_pop = pd.Series(data=data, index=index_tuples)

print("Series sử dụng Index bằng Tuple (Cách thủ công):")
vn_pop
# Khi chạy trong môi trường như Jupyter/Colab, chỉ cần gọi tên biến 'vn_pop' để hiển thị
# Nếu chạy trong môi trường console, bạn có thể dùng print(vn_pop)
```
