```python
import numpy as np
```

```python
arr_zeros = np.zeros(10, dtype=int)
arr_zeros[4] = 1  # Vị trí thứ 5 có index = 4
print("arr_zeros:", arr_zeros)
```

```python
arr_h = np.arange(10, 25)
arr_h_reversed = arr_h[::-1]
print("arr_h đảo ngược:", arr_h_reversed)
```

Cho arr_k, tạo arr_1 chỉ chứa các phần tử khác 0


```python
arr_k = np.array([1, 2, 0, 8, 2, 0, 1, 3, 0, 5, 0, 1])
arr_1 = arr_k[arr_k != 0]
print("arr_1 (khác 0):", arr_1)
```

 Thêm 2 phần tử 10 và 20 vào cuối arr_1


```python
arr_2 = np.append(arr_1, [10, 20])
print("arr_2 (thêm 10,20):", arr_2)
```

Thêm phần tử 100 vào vị trí index = 5


```python
arr_3 = np.insert(arr_2, 5, 100)
print("arr_3 (thêm 100 tại index 5):", arr_3)
```

 Xóa các phần tử tại vị trí index = 0,1,2

```python
arr_4 = np.delete(arr_3, [0, 1, 2])
print("arr_4 (xóa index 0,1,2):", arr_4)
```
