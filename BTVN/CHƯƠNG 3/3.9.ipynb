```python
import pandas as pd

# Đọc dữ liệu
euro12 = pd.read_csv("euro2012.csv")

# In euro12: type, shape, columns
print(type(euro12))
print(euro12.shape)
print(euro12.columns)
```

```python
#1, in giá trị cột Goals
print(euro12["Goals"])
```

```python
#2, số đội tham gia Eruo 2012 là:
print("Số đội:", euro12["Team"].nunique())
```

```python
#3, in thông tin

print(euro12.info())
```

```python
#4, Tạo data frame discipline (Team, Yellow Cards, Red Cards)
discipline = euro12[["Team", "Yellow Cards", "Red Cards"]]
print(discipline)
```

```python
#5. Sắp xếp discipline giảm dần theo Red Cards, Yellow Cards
discipline_sorted = discipline.sort_values(
    by=["Red Cards", "Yellow Cards"], 
    ascending=False
)
print(discipline_sorted)
```

```python
#6a. Tính trung bình Yellow Card
print("Trung bình Yellow Cards:", euro12["Yellow Cards"].mean())
```

```python
#6b. Lọc các đội ghi hơn 6 bàn thắng
print(euro12[euro12["Goals"] > 6])
```

```python
#7. In các đội bắt đầu bằng chữ 'G'
print(euro12[euro12["Team"].str.startswith("G")])
```

```python
#8. In 7 cột đầu của euro12
print(euro12.iloc[:, :7])
```

```python
#9. In tất cả các cột, trừ 3 cột 
print(euro12.iloc[:, :-3])
```

```python
#10. In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print(euro12[["Team", "Goals", "Shooting Accuracy", "Yellow Cards", "Red Cards"]])
```

```python
#11. In các cột 'Team', 'Shooting Accuracy' cho England, Italy, Russia
print(euro12[euro12["Team"].isin(["England", "Italy", "Russia"])]
      [["Team", "Shooting Accuracy"]])
```
