```python
import numpy as np
height = [74, 70, 72, 73, 69, 75, 71, 68, 70, 72, 74, 69, 71, 70, 72, 73, 74, 70, 71, 72,
          73, 74, 69, 70, 71, 72, 73, 74, 70, 71, 72, 73, 69, 70, 71, 72, 73, 74, 70, 71,
          72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71,
          72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71,
          72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71, 72, 73, 74, 70, 71]

weight = [180, 165, 190, 200, 155, 210, 175, 160, 170, 185, 195, 158, 178, 168, 182, 192, 205, 173, 180, 188,
          198, 202, 159, 172, 176, 186, 196, 208, 170, 175, 182, 193, 160, 168, 177, 185, 195, 205, 172, 180,
          187, 194, 202, 170, 175, 182, 192, 200, 172, 178, 185, 195, 203, 175, 180, 185, 193, 200, 172, 178,
          184, 192, 200, 170, 176, 182, 190, 198, 172, 178, 184, 192, 200, 170, 176, 182, 190, 198, 172, 178,
          184, 192, 200, 170, 176, 182, 190, 198, 172, 178, 184, 192, 200, 170, 176, 182, 190, 198, 172, 178]
```

```python
# Bước 1: Tạo numpy array từ list


arr_height = np.array(height)
arr_weight = np.array(weight)
```

```python
# Bước 2: Quy đổi chiều cao sang mét

inch_to_m = 0.0254
arr_height_m = arr_height * inch_to_m
```

```python
# Bước 3: Quy đổi cân nặng sang kg

pound_to_kg = 0.453592
arr_weight_kg = arr_weight * pound_to_kg

```

```python
# Bước 4: Tính BMI
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
```

```python
# Bước 5: Cân nặng tại vị trí index = 50
weight_index_50 = arr_weight_kg[50]
print("Cân nặng của cầu thủ ở vị trí index 50 (kg):", weight_index_50)
```

```python
# Bước 6: Chiều cao từ index 100 đến 110

height_100_110 = arr_height_m[100:111]
print("Chiều cao các cầu thủ từ index 100 đến 110 (m):", height_100_110)
```

```python
# Bước 7: Cầu thủ có BMI < 21
bmi_under_21 = arr_bmi[arr_bmi < 21]
print("Các cầu thủ có BMI < 21:", bmi_under_21)
```

```python
# Bước 8: Chiều cao và cân nặng trung bình

avg_height = np.mean(arr_height_m)
avg_weight = np.mean(arr_weight_kg)
print("Chiều cao trung bình (m):", avg_height)
print("Cân nặng trung bình (kg):", avg_weight)
```

```python
# Bước 9: Chiều cao và cân nặng lớn nhất
max_height = np.max(arr_height_m)
max_weight = np.max(arr_weight_kg)
print("Chiều cao lớn nhất (m):", max_height)
print("Cân nặng lớn nhất (kg):", max_weight)
```

```python
# Bước 11 / Ý 11: Chiều cao và cân nặng nhỏ nhất
min_height = np.min(arr_height_m)
min_weight = np.min(arr_weight_kg)
print("Chiều cao nhỏ nhất (m):", min_height)
print("Cân nặng nhỏ nhất (kg):", min_weight)
```
