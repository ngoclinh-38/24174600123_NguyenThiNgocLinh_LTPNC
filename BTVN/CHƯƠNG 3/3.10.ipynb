```python
import pandas as pd
```

```python
drink = pd.read_csv("drinks.csv", index_col=0)
print(type(drink))
print(drink.shape)
print(drink.columns)
print(drink.head())
print(drink.tail())
```

```python
# 2. Trung bình bia theo châu lục
print(drink.groupby("continent")["beer_servings"].mean())
```

```python
# 3. Thống kê rượu vang theo châu lục
print(drink.groupby("continent")["wine_servings"].describe())
```

```python
# 4. Mean bia + rượu theo châu lục
print(drink.groupby("continent")[["beer_servings", "wine_servings"]].mean())
```

```python
# 5. Median bia + rượu theo châu lục
print(drink.groupby("continent")[["beer_servings", "wine_servings"]].median())
```

```python
# 6. Spirit: mean – max – min theo châu lục
print(drink.groupby("continent")["spirit_servings"].agg(["mean", "max", "min"]))
```

```python
# 7. Sắp xếp theo beer_servings

sorted_beer = drink.sort_values(by="beer_servings")
print(sorted_beer)

# 5 ít nhất
print(sorted_beer.head(5))

# 5 nhiều nhất
print(sorted_beer.tail(5))
```
