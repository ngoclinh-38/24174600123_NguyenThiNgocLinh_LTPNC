```python
import numpy as np
```

```python
arr = np.arange(10)  # Tạo mảng 
print("Các phần tử của arr:", arr)
print("Kiểu dữ liệu của arr:", arr.dtype)
print("Kích thước của arr:", arr.shape)
```

```python
# Tạo arr_odd và arr_even
arr_odd = arr[arr % 2 == 1]    # Lọc các số lẻ
arr_even = arr[arr % 2 == 0]   # Lọc các số chẵn
print("Các phần tử lẻ (arr_odd):", arr_odd)
print("Các phần tử chẵn (arr_even):", arr_even)
```

```python
#  Tạo arr_update1 các phần tử chẵn giữ nguyên, phần tử lẻ thay bằng 100
arr_update1 = np.where(arr % 2 == 0, arr, 100)
print("arr_update1:", arr_update1)
```
