Viết chương trình Python để tạo kết nối cơ sở dữ liệu SQLite tới cơ sở dữ liệu nằm trong bộ nhớ.

```python
import sqlite3

# Tạo kết nối tới CSDL SQLite trong bộ nhớ
conn = sqlite3.connect(":memory:")

# Tạo con trỏ
cursor = conn.cursor()

print("Đã kết nối tới cơ sở dữ liệu SQLite trong bộ nhớ")

# Đóng kết nối
conn.close()
```
