```python
import sqlite3

def delete_student_by_name(db_name, student_name):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    cursor.execute(
        "DELETE FROM Student WHERE name = ?",
        (student_name,)
    )

    conn.commit()
    conn.close()

# Gọi hàm
delete_student_by_name("student.db", "An")
```
