Viết chương trình tạo một cơ sở dữ liệu SQLite và tạo một bảng nằm trong cơ sở dữ liệu.

```python
import sqlite3
# Tạo và kết nối tới cơ sở dữ liệu SQLite
conn = sqlite3.connect("student.db")

# Tạo con trỏ
cursor = conn.cursor()

# Tạo bảng Student
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

print("Tạo cơ sở dữ liệu và bảng Student thành công")

# Lưu thay đổi và đóng kết nối
conn.commit()
conn.close()
```
