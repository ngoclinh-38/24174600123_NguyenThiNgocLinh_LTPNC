```python
import sqlite3
import pandas as pd

# Kết nối tới CSDL SQLite
conn = sqlite3.connect("student.db")

# Đọc dữ liệu từ bảng vào DataFrame
df = pd.read_sql("SELECT * FROM Student", conn)

# Đếm số bản ghi
count = len(df)

print("Số bản ghi trong bảng Student:", count)

# Đóng kết nối
conn.close()
```
