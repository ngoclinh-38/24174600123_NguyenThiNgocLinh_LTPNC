Viết chương trình Python để tạo một cơ sở dữ liệu SQLite, kết nối với cơ sở dữ liệu và in ra version của cơ sở dữ liệu SQLite.

```python
import sqlite3

# Tạo và kết nối tới cơ sở dữ liệu SQLite (tạo mới nếu chưa có)
conn = sqlite3.connect("example.db")

# Tạo con trỏ
cursor = conn.cursor()

# Lấy version của SQLite
cursor.execute("SELECT sqlite_version();")

# In ra version
version = cursor.fetchone()
print("SQLite version:", version[0])

# Đóng kết nối
conn.close()
```
