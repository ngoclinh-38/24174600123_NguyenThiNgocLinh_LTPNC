Tạo dữ liệu mẫu

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("product.db")

df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Pen", "Book", "Bag"],
    "price": [10.5, 25.0, 120.0],
    "amount": [100, 50, 20]
})

df.to_sql("product", conn, if_exists="replace", index=False)
```

Hiển thị danh sách sản phẩm

```python
df = pd.read_sql("SELECT * FROM product", conn)
print(df)
```

Thêm sản phẩm mới


```python
new_product = pd.DataFrame({
    "id": [4],
    "name": ["Notebook"],
    "price": [45.0],
    "amount": [30]
})

df = pd.concat([df, new_product], ignore_index=True)
df.to_sql("product", conn, if_exists="replace", index=False)
```

Tìm kiếm sản phẩm theo tên

```python
result = df[df["name"].str.contains("Book")]
print(result)
```

Cập nhật giá và số lượng theo id


```python
df.loc[df["id"] == 2, ["price", "amount"]] = [30.0, 60]
df.to_sql("product", conn, if_exists="replace", index=False)
```

Xoá sản phẩm theo id

```python
df = df[df["id"] != 1]
df.to_sql("product", conn, if_exists="replace", index=False)
```

Sắp xếp theo giá tăng dần


```python
sorted_df = df.sort_values(by="price")
print(sorted_df)
```

Lưu kết quả cuối cùng

```python
sorted_df.to_sql("product", conn, if_exists="replace", index=False)
conn.close()
```
