```python
import threading
import time

N = 20  # ngưỡng
```

Luồng ghi số chẵn

```python
def ghi_so_chan():
    print("Các số chẵn:")
    for i in range(0, N + 1, 2):
        print(i, end=" ")
        time.sleep(0.1)
    print()
```

Luồng ghi số lẻ

```python
def ghi_so_le():
    print("Các số lẻ:")
    for i in range(1, N + 1, 2):
        print(i, end=" ")
        time.sleep(0.1)
    print()
```

Tạo luồng

```python
t1 = threading.Thread(target=ghi_so_chan)
t2 = threading.Thread(target=ghi_so_le)
```

Chạy luồng

```python
t1.start()
t2.start()
```

```python
t1.join()
t2.join()

print("Kết thúc chương trình")
```
