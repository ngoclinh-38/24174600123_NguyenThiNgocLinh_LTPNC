Thực hiện các yêu cầu HTTP đồng thời bằng luồng

```python
import threading
import requests

# Hàm gửi yêu cầu HTTP
def gui_yeu_cau(url):
    print(f"Đang gửi yêu cầu tới: {url}")
    response = requests.get(url)
    print(f"Hoàn thành: {url} | Mã trạng thái: {response.status_code}")

# Danh sách các URL
urls = [
    "https://www.example.com",
    "https://www.python.org",
    "https://www.wikipedia.org"
]

threads = []

# Tạo và chạy các luồng
for url in urls:
    t = threading.Thread(target=gui_yeu_cau, args=(url,))
    threads.append(t)
    t.start()

# Chờ các luồng hoàn thành
for t in threads:
    t.join()

print("Đã hoàn thành tất cả các yêu cầu HTTP")
```
