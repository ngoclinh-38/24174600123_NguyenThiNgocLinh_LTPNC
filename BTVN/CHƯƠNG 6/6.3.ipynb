Tạo 2 luồng để tìm và in số chẵn, số lẻ từ 30 đến 50

```python
import threading

# Luồng in số chẵn
def in_so_chan():
    print("Các số chẵn:")
    for i in range(30, 51):
        if i % 2 == 0:
            print(i, end=" ")
    print()

# Luồng in số lẻ
def in_so_le():
    print("Các số lẻ:")
    for i in range(30, 51):
        if i % 2 != 0:
            print(i, end=" ")
    print()

# Tạo 2 luồng
t1 = threading.Thread(target=in_so_chan)
t2 = threading.Thread(target=in_so_le)

# Chạy luồng
t1.start()
t2.start()

# Chờ 2 luồng hoàn thành
t1.join()
t2.join()

print("Chương trình kết thúc")
```
