Tải nhiều tệp đồng thời bằng luồng

```python
import threading
import urllib.request

# Hàm tải file
def tai_file(url, ten_file):
    print(f"Đang tải: {ten_file}")
    urllib.request.urlretrieve(url, ten_file)
    print(f"Tải xong: {ten_file}")

# Danh sách các file cần tải
files = [
    ("https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf", "file1.pdf"),
    ("https://www.learningcontainer.com/wp-content/uploads/2020/05/sample-text-file.txt", "file2.txt"),
    ("https://www.learningcontainer.com/wp-content/uploads/2020/04/sample-img.png", "file3.png")
]

threads = []

# Tạo và chạy các luồng
for url, ten in files:
    t = threading.Thread(target=tai_file, args=(url, ten))
    threads.append(t)
    t.start()

# Chờ các luồng hoàn thành
for t in threads:
    t.join()

print("Đã tải xong tất cả các tệp")
```
