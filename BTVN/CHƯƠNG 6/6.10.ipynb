Dùng Thread tìm giá trị lớn nhất trong list

```python
import threading
import random
```

Hàm tìm mas của list con

```python
def tim_max_list_con(list_con, index, ket_qua):
    max_con = max(list_con)
    print(f"Thread {index}: {list_con} -> Max = {max_con}")
    ket_qua[index] = max_con
```

Chương trình chính

```python
n = int(input("Nhập số phần tử: "))
so_thread = int(input("Nhập số thread: "))
```

Tạo list ngẫu nhiên từ 0 đến 100

```python
lst = [random.randint(0, 100) for _ in range(n)]
print("List ban đầu:", lst)
```

Chia list cho các Thread

```python
kich_thuoc = n // so_thread
threads = []
ket_qua = [0] * so_thread

bat_dau = 0

for i in range(so_thread):
    if i == so_thread - 1:
        list_con = lst[bat_dau:]
    else:
        list_con = lst[bat_dau: bat_dau + kich_thuoc]

    t = threading.Thread(
        target=tim_max_list_con,
        args=(list_con, i + 1, ket_qua)
    )
    threads.append(t)
    t.start()

    bat_dau += kich_thuoc
```

Các Thread hoàn thành

```python
for t in threads:
    t.join()

# Tìm max của list lớn
max_list = max(ket_qua)
print("Giá trị lớn nhất trong list =", max_list)
```
