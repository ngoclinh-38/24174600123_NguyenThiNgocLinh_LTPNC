Dùng Thread tính tổng các phần tử trong list

```python
import threading
import random
```

Hàm tính tổng các list con

```python
def tinh_tong_list_con(list_con, index, ket_qua):
    tong = sum(list_con)
    print(f"Thread {index}: {list_con} -> Tổng = {tong}")
    ket_qua[index] = tong
```

Chương trình chính

```python
n = int(input("Nhập số phần tử: "))
so_thread = int(input("Nhập số thread: "))
```

Tạo list ngẫu nhiên từ 0-10

```python
lst = [random.randint(0, 10) for _ in range(n)]
print("List ban đầu:", lst)
```

Chia list thành các phần cho Thread

```python
kich_thuoc = n // so_thread
threads = []
ket_qua = [0] * so_thread

bat_dau = 0

for i in range(so_thread):
    if i == so_thread - 1:
        list_con = lst[bat_dau:]
    else:
        list_con = lst[bat_dau: bat_dau + kich_thuoc]

    t = threading.Thread(
        target=tinh_tong_list_con,
        args=(list_con, i + 1, ket_qua)
    )
    threads.append(t)
    t.start()

    bat_dau += kich_thuoc
```

Thread hoàn thành và tổng list lớn

```python
for t in threads:
    t.join()

tong_list = sum(ket_qua)
print("Tổng list =", tong_list)
```
