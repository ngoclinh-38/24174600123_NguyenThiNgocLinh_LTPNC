Ứng dụng multithread (Google – Yahoo – Facebook)

```python
import threading
import time
```

Hàm chạy cho mỗi luồng

```python
def hien_thi(ten, delay, so_lan):
    print(f"Starting {ten}")
    for i in range(so_lan):
        time.sleep(delay)
        print(f"{ten}: {time.asctime(time.localtime())}")
    print(f"Exiting {ten}")
```

Main Thread

```python
print("Starting Main Thread")
```

Tạo các luồng

```python
t1 = threading.Thread(target=hien_thi, args=("Google", 1, 4))
t2 = threading.Thread(target=hien_thi, args=("Yahoo", 2, 3))
t3 = threading.Thread(target=hien_thi, args=("Facebook", 3, 3))
```

Chạy các luồng

```python
t1.start()
t2.start()
t3.start()
```

Các luồng hàon thành

```python
t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```
