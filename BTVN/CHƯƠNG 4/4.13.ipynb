```python
import pandas as pd
import numpy as np
```

```python
ser1 = pd.Series([10, 9, 6, 5, 3, 1, 12, 8, 13])
ser2 = pd.Series([1, 3, 10, 13])
```

Xếp chồng theo chiều ngang (horizontal)

```python
df_horizontal = pd.concat([ser1.reset_index(drop=True), ser2.reindex(ser1.index)], axis=1)
df_horizontal.columns = ['ser1', 'ser2']
print("DataFrame xếp theo chiều ngang:")
print(df_horizontal)
```

Xếp chồng theo chiều dọc (vertical)


```python
df_vertical = pd.concat([ser1, ser2], axis=0).reset_index(drop=True)
print("\nSeries xếp theo chiều dọc:")
print(df_vertical)
```

Lấy vị trí của các mục của ser2 trong ser1

```python
positions = [ser1[ser1 == val].index.tolist() for val in ser2]
print("\nVị trí của các giá trị của ser2 trong ser1:")
print(positions)
```
