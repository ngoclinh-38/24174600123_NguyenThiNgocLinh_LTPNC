```python
import pandas as pd
```

1a. Đọc stocks1.csv


```python
stocks1 = pd.read_csv("stocks1.csv")
print("5 dòng đầu stocks1:\n", stocks1.head())
print("5 dòng cuối stocks1:\n", stocks1.tail())
print("Kiểu dữ liệu cột stocks1:\n", stocks1.dtypes)
print("Info stocks1:")
stocks1.info()
```


1b. Đọc stocks2.csv


```python
stocks2 = pd.read_csv("stocks2.csv")
print("\n5 dòng đầu stocks2:\n", stocks2.head())
print("5 dòng cuối stocks2:\n", stocks2.tail())
print("Kiểu dữ liệu cột stocks2:\n", stocks2.dtypes)
print("Info stocks2:")
stocks2.info()
```

1c. Đọc companies.csv


```python
companies = pd.read_csv("companies.csv")
print("\nDữ liệu companies:\n", companies)
print("Kiểu dữ liệu cột companies:\n", companies.dtypes)
print("Info companies:")
companies.info()
```


2. Kiểm tra Null và thay thế
 

```python
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
```

3. Gộp stocks1 và stocks2 theo dòng


```python
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print("\n15 dòng cuối stocks:")
print(stocks.tail(15))
```

4. Gộp stocks và companies


```python

stocks_companies = pd.merge(stocks, companies, left_on='symbol', right_on='name', how='left')
print("\n5 dòng đầu stocks_companies:")
print(stocks_companies.head())
```


5. Giá trung bình (open, high, low, close) và volume trung bình của mỗi công ty

```python
mean_stats = stocks_companies.groupby('symbol')[['open','high','low','close','volume']].mean()
print("\nGiá trung bình và volume trung bình mỗi công ty:")
print(mean_stats)
```

6. Giá đóng cửa (close) trung bình, lớn nhất, nhỏ nhất

```python
close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean','max','min'])
print("\nClose trung bình, max, min mỗi công ty:")
print(close_stats)
```


7. Tạo cột parsed_time

```python
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("\nKiểu dữ liệu parsed_time:", stocks_companies['parsed_time'].dtype)
print("5 dòng đầu stocks_companies với parsed_time:")
print(stocks_companies.head())
```


8. Thêm cột result

```python
stocks_companies['result'] = stocks_companies.apply(lambda x: 'up' if x['close'] > x['open'] else 'down', axis=1)
print("\n5 dòng đầu stocks_companies với cột result:")
print(stocks_companies.head())
```
