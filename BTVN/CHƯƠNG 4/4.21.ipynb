```python
import pandas as pd
```

1. Đọc dữ liệu từ file CSV

```python
drink = pd.read_csv("drinks.csv", index_col=0)
# Thông tin cơ bản
print("Type:", type(drink))
print("Shape:", drink.shape)
print("Columns:", drink.columns.tolist())
print("\n5 dòng đầu:")
print(drink.head())
print("\n5 dòng cuối:")
print(drink.tail())
```


2. Số lượng bia tiêu thụ trung bình theo châu lục


```python
beer_mean_by_continent = drink.groupby('continent')['beer_servings'].mean()
print("\nSố lượng bia tiêu thụ trung bình theo châu lục:")
print(beer_mean_by_continent)
```

3. Thống kê rượu vang theo châu lục


```python

wine_stats_by_continent = drink.groupby('continent')['wine_servings'].describe()
print("\nThông tin thống kê rượu vang theo châu lục:")
print(wine_stats_by_continent)
```

4. Trung bình bia và rượu theo châu lục


```python
mean_beer_spirit = drink.groupby('continent')[['beer_servings','spirit_servings']].mean()
print("\nTrung bình bia và rượu theo châu lục:")
print(mean_beer_spirit)
```

5. Trung vị bia và rượu theo châu lục

```python
median_beer_spirit = drink.groupby('continent')[['beer_servings','spirit_servings']].median()
print("\nTrung vị bia và rượu theo châu lục:")
print(median_beer_spirit)
```

6. Spirit servings: trung bình, lớn nhất, nhỏ nhất

```python
spirit_stats = drink.groupby('continent')['spirit_servings'].agg(['mean','max','min'])
print("\nSpirit servings theo châu lục:")
print(spirit_stats)
```

7. Sắp xếp dữ liệu theo beer_servings tăng dần

```python
sorted_drink = drink.sort_values('beer_servings')
print("\n5 quốc gia tiêu thụ bia ít nhất:")
print(sorted_drink.head())
print("\n5 quốc gia tiêu thụ bia nhiều nhất:")
print(sorted_drink.tail())
```
