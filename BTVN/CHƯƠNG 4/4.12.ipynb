```python
import pandas as pd
```

```python
ser1 = pd.Series(range(5))
ser2 = pd.Series(list('abcde'))
```

```python
# 1. Xếp chồng theo chiều ngang (tạo DataFrame 5x2)
df_horizontal = pd.concat([ser1, ser2], axis=1)
df_horizontal.columns = ['Numbers', 'Letters']
print("DataFrame xếp theo chiều ngang:")
print(df_horizontal)
```

```python
# 2. Xếp chồng theo chiều dọc (tạo Series dài 10)
df_vertical = pd.concat([ser1, ser2], axis=0).reset_index(drop=True)
print("\nSeries xếp theo chiều dọc:")
print(df_vertical)
```
