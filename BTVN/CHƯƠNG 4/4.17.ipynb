```python
import pandas as pd

ser = pd.Series([
    '01 Jan 2010',
    '02-02-2011',
    '20120303',
    '2013/04/04',
    '2014-05-05',
    '2015-06-06T12:20'
])
```

Chuyển Series sang datetime, nếu có giá trị không hợp lệ sẽ trả NaT

```python
ser_datetime = pd.to_datetime(ser, errors='coerce')

print("Series sau khi chuyển sang datetime:")
print(ser_datetime)
```
