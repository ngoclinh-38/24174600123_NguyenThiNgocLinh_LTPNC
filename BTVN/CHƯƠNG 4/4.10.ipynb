```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
ser = pd.Series(np.random.randint(1, 10, 7))
print("Original Series:")
print(ser)
```

Tìm vị trí các số là bội của 3


```python
indices_multiple_of_3 = ser[ser % 3 == 0].index
print("\nVị trí các số là bội của 3:")
print(indices_multiple_of_3.tolist())
```
