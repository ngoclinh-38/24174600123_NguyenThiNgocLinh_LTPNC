```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
ser = pd.Series(np.random.randint(1, 10, 35))
print("Original Series:")
print(ser)

# Giải thích:
"""
- np.random.randint(1, 10, 35) → tạo 1 mảng gồm 35 số nguyên ngẫu nhiên từ 1 đến 9 (10 không tính)
- pd.Series(...) → biến mảng này thành Pandas Series
"""
```

Định hình lại Series thành DataFrame 7 hàng x 5 cột


```python
df = ser.values.reshape(7, 5)  # reshape mảng numpy
df = pd.DataFrame(df, columns=[f'Col{i+1}' for i in range(5)])
print("\nDataFrame 7 hàng x 5 cột:")
print(df)
```
