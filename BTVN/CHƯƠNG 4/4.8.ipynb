```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
ser = pd.Series(np.random.random(20))
print("Original Series:")
print(ser)
```

Tạo nhãn phân vị: 10 deciles (0-10%, 10-20%, ..., 90-100%)


```python
decile_labels = [f'D{i+1}' for i in range(10)]
```

Chia series thành 10 phân vị và thay bằng nhãn


```python
ser_deciles = pd.qcut(ser, 10, labels=decile_labels)
print("\nSeries sau khi chuyển thành deciles:")
print(ser_deciles)
```
