```python
import pandas as pd
import numpy as np

np.random.seed(42)  # để kết quả giống nhau
```

```python
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))

print("Fruit:", fruit.tolist())
print("Weights:", weights.tolist())
```

```python
# Tính trọng số trung bình cho từng loại quả
weighted_avg = fruit.groupby(fruit).apply(lambda x: weights[x.index].mean())
print("\nTrọng số trung bình của từng loại quả:")
print(weighted_avg)
```
