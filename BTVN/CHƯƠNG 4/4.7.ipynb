```python
import pandas as pd
import numpy as np

```

```python
rng = np.random.RandomState(100)  # tạo random seed
ser = pd.Series(rng.randint(1, 5, 12))
print("Original Series:")
print(ser)
```

Tìm top 2 giá trị phổ biến nhất


```python
top2 = ser.value_counts().nlargest(2).index
print("\nTop 2 most frequent values:", list(top2))
```

Thay các giá trị khác bằng 'Other'

```python
ser_modified = ser.where(ser.isin(top2), 'Other')
print("\nSeries sau khi giữ top 2 và thay các giá trị khác bằng 'Other':")
print(ser_modified)
```
