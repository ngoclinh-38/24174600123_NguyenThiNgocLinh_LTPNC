```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)  # để kết quả lặp lại
ser = pd.Series(np.take(list('abcdefgh'), np.random.randint(8, size=30)))

print("Series:")
print(ser)
```

```python
"""
- list('abcdefgh') → tạo danh sách ['a','b','c','d','e','f','g','h']
- np.random.randint(8, size=30) → tạo 1 mảng 30 số nguyên ngẫu nhiên từ 0 đến 7
- np.take(list(...), indices) → lấy các phần tử theo chỉ số của mảng indices
  => tạo ra 1 mảng 30 phần tử ngẫu nhiên từ 'a' đến 'h'
- pd.Series(...) → biến mảng thành Series
"""
```

Tính số lần xuất hiện của mỗi giá trị duy nhất


```python
counts = ser.value_counts()
print("\nSố lần xuất hiện mỗi giá trị duy nhất:")
print(counts)
```
