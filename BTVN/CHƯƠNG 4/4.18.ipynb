```python
import pandas as pd
```

Input: Series chứa nhiều chuỗi, một số là email hợp lệ


```python
emails = pd.Series([
    'buying books at amazom.com',
    'rameses@egypt.com',
    'matt@t.co',
    'narendra@modi.com'
])
```

Regex chuẩn để kiểm tra email


```python
pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}'
```

Lọc các email hợp lệ

```python
valid_emails = emails[emails.str.contains(pattern, regex=True)]
print("Các email hợp lệ:")
print(valid_emails.reset_index(drop=True))
```
