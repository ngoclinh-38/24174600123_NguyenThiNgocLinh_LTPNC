```python
import pandas as pd
import numpy as np
```

```python
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict)
```

 Chuyển Series thành DataFrame


```python
df = ser.reset_index()
df.columns = ['letter', 'number']  # Đổi tên cột

print(df.head())
```
