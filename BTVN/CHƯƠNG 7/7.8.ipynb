Bài 7.8a – Liệt kê các ước của N (console)


```python
n = int(input("Enter N: "))

print("The divisors of N are:", end=" ")
i = 1
while i <= n:
    if n % i == 0:
        print(i, end=" ")
    i += 1
```

BÀI 7.8b

```python
import tkinter as tk

# Hàm xử lý khi bấm Validate
def find_divisors():
    try:
        n = int(entry.get())
        divisors = ""
        i = 1
        while i <= n:
            if n % i == 0:
                divisors += str(i) + " "
            i += 1

        result_label.config(
            text="The divisors of N : " + divisors
        )
    except:
        result_label.config(text="Please enter a valid integer")

# Tạo cửa sổ
root = tk.Tk()
root.title("tk")
root.geometry("500x200")

# Nhãn nhập
label = tk.Label(root, text="Enter the value of N")
label.pack(pady=10)

# Ô nhập
entry = tk.Entry(root, width=20)
entry.pack()

# Nhãn kết quả
result_label = tk.Label(root, text="")
result_label.pack(pady=10)

# Nút Validate
button = tk.Button(root, text="Validate", command=find_divisors)
button.pack(pady=10)

# Chạy chương trình
root.mainloop()
```
