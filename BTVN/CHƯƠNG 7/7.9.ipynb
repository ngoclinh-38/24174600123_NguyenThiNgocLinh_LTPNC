```python
import tkinter as tk
from tkinter import ttk

# Hàm tính GCD (không dùng math.gcd)
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# Hàm xử lý khi chọn chức năng
def calculate():
    try:
        m = int(entry_m.get())
        n = int(entry_n.get())
        func = combo.get()

        if func == "GCD":
            result = gcd(m, n)
            result_label.config(text=f"GCD(m, n) = {result}")
        else:
            result = (m * n) // gcd(m, n)
            result_label.config(text=f"LCM(m, n) = {result}")
    except:
        result_label.config(text="Please enter valid integers")

# Tạo cửa sổ
root = tk.Tk()
root.title("tk")
root.geometry("500x250")

# Nhập m
tk.Label(root, text="Enter value of m:").pack(pady=5)
entry_m = tk.Entry(root)
entry_m.pack()

# Nhập n
tk.Label(root, text="Enter value of n:").pack(pady=5)
entry_n = tk.Entry(root)
entry_n.pack()

# Chọn chức năng
tk.Label(root, text="Choose function:").pack(pady=5)
combo = ttk.Combobox(root, values=["GCD", "LCM"])
combo.current(0)
combo.pack()

# Nút Validate
tk.Button(root, text="Validate", command=calculate).pack(pady=10)

# Hiển thị kết quả
result_label = tk.Label(root, text="")
result_label.pack(pady=10)

# Chạy chương trình
root.mainloop()
```
