```python
import tkinter as tk
from tkinter import messagebox
```

HÀm hiển thị hộp thoại

```python
def show_exercise(title, content):
    messagebox.showinfo(title, content)
```

Tạo menu cho một chương

```python
def create_chapter_menu(menu_bar, chapter_number):
    chapter_menu = tk.Menu(menu_bar, tearoff=0)

    for i in range(1, 10):
        if chapter_number == 1 and i == 1:
            chapter_menu.add_command(
                label=f"{chapter_number}.{i}",
                command=lambda: show_exercise(
                    "Lập trình hướng đối tượng",
                    "Giải phương trình bậc hai"
                )
            )
        else:
            chapter_menu.add_command(
                label=f"{chapter_number}.{i}",
                command=lambda m=chapter_number, n=i: show_exercise(
                    "Bài tập",
                    f"Bài tập {m}.{n}"
                )
            )

    chapter_menu.add_separator()
    chapter_menu.add_command(label="Exit", command=root.quit)

    menu_bar.add_cascade(label=f"Chương {chapter_number}", menu=chapter_menu)
```

tạo cửa sổ và menu

```python
root = tk.Tk()
root.title("Hệ thống bài tập nâng cao Python")
root.geometry("600x400")

menu_bar = tk.Menu(root)
root.config(menu=menu_bar)

for chapter in range(1, 8):
    create_chapter_menu(menu_bar, chapter)
```

Chạy chương trình

```python
root.mainloop()
```
