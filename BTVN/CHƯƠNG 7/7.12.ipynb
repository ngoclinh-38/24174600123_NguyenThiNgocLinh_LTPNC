TÍNH CHỈ SỐ BMI

```python
import tkinter as tk
from tkinter import messagebox
```

Hàm tính BMI

```python
def calculate_bmi():
    global entry_weight, entry_height, result_label, status_label
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())

        if height <= 0 or weight <= 0:
            raise ValueError

        bmi = weight / (height ** 2)

        # Phân loại BMI
        if bmi < 18.5:
            status = "Gầy"
        elif bmi < 25:
            status = "Bình thường"
        elif bmi < 30:
            status = "Thừa cân"
        else:
            status = "Béo phì"

        result_label.config(text=f"BMI = {bmi:.2f}")
        status_label.config(text=f"Tình trạng: {status}")

    except:
        messagebox.showerror("Lỗi", "Vui lòng nhập cân nặng và chiều cao hợp lệ")
```

Tạo giao diện

```python
root = tk.Tk()
root.title("Tính chỉ số BMI")
root.geometry("400x300")

tk.Label(root, text="Cân nặng (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack(pady=5)
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=calculate_bmi).pack(pady=15)

result_label = tk.Label(root, text="", font=("Arial", 14), fg="blue")
result_label.pack(pady=10)

status_label = tk.Label(root, text="", font=("Arial", 12))
status_label.pack()
```

Chạy chương trình

```python
root.mainloop()
```
