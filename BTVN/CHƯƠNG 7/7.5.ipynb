```python
from PIL import Image, ImageDraw

# Tạo ảnh mới
img = Image.new("RGB", (400, 200), "white")
draw = ImageDraw.Draw(img)

# Vẽ ô tô đơn giản
draw.rectangle([80, 90, 320, 140], fill="blue")     # thân xe
draw.rectangle([140, 60, 260, 90], fill="blue")    # mui xe
draw.ellipse([110, 130, 150, 170], fill="black")   # bánh trước
draw.ellipse([250, 130, 290, 170], fill="black")   # bánh sau

# Lưu thành file Otto.png
img.save("Otto.png")

print("Đã tạo file Otto.png")
```
