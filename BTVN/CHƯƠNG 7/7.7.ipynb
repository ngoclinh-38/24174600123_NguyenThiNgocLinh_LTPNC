```python
import tkinter as tk

# Hàm xử lý khi bấm nút Validate
def calculate_sum():
    try:
        n = int(entry.get())
        total = 0
        i = 1
        while i <= n:
            total += i
            i += 1
        
        result_label.config(
            text=f"The sum is 1 + 2 + ... + {n} = {total}"
        )
    except:
        result_label.config(text="Please enter a valid integer")

# Tạo cửa sổ
root = tk.Tk()
root.title("tk")
root.geometry("500x200")

# Nhãn nhập
label = tk.Label(root, text="Enter value of integer N :")
label.pack(pady=10)

# Ô nhập
entry = tk.Entry(root, width=20)
entry.pack()

# Nhãn kết quả
result_label = tk.Label(root, text="")
result_label.pack(pady=10)

# Nút Validate
button = tk.Button(root, text="Validate", command=calculate_sum)
button.pack(pady=10)

# Chạy chương trình
root.mainloop()
```
