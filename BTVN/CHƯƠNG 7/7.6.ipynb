# Viết chương trình nhập một chuỗi và trả lại chuỗi đảo ngược

```python
# KHÔNG sử dụng các hàm đảo chuỗi có sẵn
# Nhập chuỗi
s = input("Enter a word: ")

# Đảo chuỗi thủ công
result = ""
i = len(s) - 1

while i >= 0:
    result = result + s[i]
    i = i - 1

# Xuất kết quả
print("Reversed string:", result)
```
