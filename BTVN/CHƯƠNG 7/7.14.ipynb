```python
import tkinter as tk
from tkinter import ttk, messagebox
```

Hàm tính tiền taxi

```python
def calculate_fare():
    global entry_km, entry_wait, car_type, result_label
    try:
        km = float(entry_km.get())
        wait = int(entry_wait.get())
        car = car_type.get()

        # Tiền mở cửa
        total = 12000

        # Tính tiền quãng đường
        if car == "4 chỗ":
            if km <= 0.8:
                distance_cost = 0
            elif km <= 30:
                distance_cost = (km - 0.8) * 15300
            else:
                distance_cost = (30 - 0.8) * 15300 + (km - 30) * 12100
        else:  # 7 chỗ
            if km <= 0.8:
                distance_cost = 0
            elif km <= 30:
                distance_cost = (km - 0.8) * 16100
            else:
                distance_cost = (30 - 0.8) * 16100 + (km - 30) * 13800

        # Tính tiền chờ
        if wait <= 5:
            wait_cost = 0
        else:
            wait_cost = (wait - 5) * 750

        total += distance_cost + wait_cost

        result_label.config(
            text=f"Tổng tiền taxi: {int(total):,} đồng"
        )

    except:
        messagebox.showerror("Lỗi", "Vui lòng nhập dữ liệu hợp lệ")
```

Tạo giao diện

```python
root = tk.Tk()
root.title("Tính cước taxi")
root.geometry("450x350")

tk.Label(root, text="Quãng đường (km):").pack(pady=5)
entry_km = tk.Entry(root)
entry_km.pack()

tk.Label(root, text="Thời gian chờ (phút):").pack(pady=5)
entry_wait = tk.Entry(root)
entry_wait.pack()

tk.Label(root, text="Loại xe:").pack(pady=5)
car_type = ttk.Combobox(root, values=["4 chỗ", "7 chỗ"])
car_type.current(0)
car_type.pack()

tk.Button(root, text="Tính tiền", command=calculate_fare).pack(pady=15)

result_label = tk.Label(root, text="", font=("Arial", 14), fg="blue")
result_label.pack(pady=15)
```

Chạy chương trình

```python
root.mainloop()
```
