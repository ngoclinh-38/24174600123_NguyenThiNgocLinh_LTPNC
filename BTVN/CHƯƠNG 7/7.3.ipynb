THAY ĐỔI PHÔNG CHỮ LABEL (Tkinter)

```python
from tkinter import *
from tkinter import font

# Tạo cửa sổ
root = Tk()
root.title("Bài 7.3 - Font Label")

# Tạo font chữ mới
my_font = font.Font(
    family="Arial",   # Tên phông chữ
    size=16,          # Kích thước
    weight="bold"     # In đậm
)

# Tạo nhãn với font mới
label = Label(root, text="Label với phông chữ tùy chỉnh", font=my_font)

# Đưa nhãn lên cửa sổ
label.pack(padx=20, pady=20)

# Hiển thị cửa sổ
root.mainloop()
```
